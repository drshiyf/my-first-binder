```python
# Import pandas and load the file into a DataFrame
import pandas as pd
```

```python
# Load the Series Matrix file into pandas
df = pd.read_csv("GSE225001_series_matrix.txt.gz", sep="\t", comment="!", compression="gzip")
```

```python
# Show the first few rows to understand its structure
df.head()
```

```python
# Check for missing values
missing_data = df.isnull().sum()
```

```python
print(missing_data)
```

```python
# Z-score normalization for each gene (row)
df_filtered = df.copy()
df_normalized = df_filtered.copy()  # Make a copy to keep the original data
df_normalized.iloc[:, 1:] = df_filtered.iloc[:, 1:].apply(lambda x: (x - x.mean()) / x.std(), axis=1)
```

```python
df_normalized.head()
```

```python
# Save the cleaned, filtered, and normalized data
df_normalized.to_csv("GSE225001_cleaned_normalized.csv", index=False)
```
