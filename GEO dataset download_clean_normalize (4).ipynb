```python
# Load the Series Matrix file into pandas
df_expression = pd.read_csv("GSE11098-GPL1261_series_matrix.txt.gz", sep="\t", comment="!", compression="gzip",index_col=0)
df_expression = df_expression.reset_index().rename(columns={"index": "ID"})  # Ensure 'ID' is a column
```

```python
df_expression.head()
```

```python
df_gpl1261_genesymbol=df_gpl1261_annotation
df_gpl1261_genesymbol.head()
```

```python
df_expression.rename(columns={"ID_REF": "ID"}, inplace=True)
```

```python
df_expression.head()
```

```python
# Merge expression data with annotation
df_merged = df_expression.merge(df_gpl1261_genesymbol, on="ID", how="left")
```

```python
df_merged.head()
```

```python
# Drop Probe ID and set Gene Symbol as index
df_merged.drop(columns=["ID"], inplace=True)
df_merged.set_index("Gene Symbol", inplace=True)
```

```python
# If multiple probes map to the same gene, take the average expression
df_final = df_merged.groupby("Gene Symbol").mean()
```

```python
df_final.head()
```

```python
# Save the Processed Data
df_final.to_csv("GSE11098-GPL1261_gene_expression.csv")
```

```python
# Filter out genes that have low expression in most samples
expression_threshold = 10  # Example threshold, adjust based on your data
df_filtered = df_final.loc[df_final.mean(axis=1) > expression_threshold]
```

```python
# Z-score normalization for each gene (row)
df_normalized = df_final.copy()  # Make a copy to keep the original data
df_normalized.iloc[:] = df_final.iloc[:].apply(lambda x: (x - x.mean()) / x.std(), axis=1)
```

```python
df_normalized.head()
```

```python
# Save the cleaned, filtered, and normalized data
df_normalized.to_csv("GSE11098-GPL1261_cleaned_normalized.csv")
```
